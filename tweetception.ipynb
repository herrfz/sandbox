```python
import twitter
import json
import urllib
import time
import requests
```

```python
api = twitter.Api(consumer_key='', 
                  consumer_secret='', 
                  access_token_key='', 
                  access_token_secret='')
```

```python
import pandas as pd

pd.set_option('display.max_colwidth', 145)

def get_tweetception(sid):
    users = []
    text = []
    i = 0

    while i < 100:
        try:
            s = api.GetStatus(id=sid)
            users.append(s.user.GetScreenName())
            text.append(s.text)
            url = s.text.split()[-1]
            r = requests.get('http://api.longurl.org/v2/expand?url={}&format=json'.format(urllib.quote_plus(url)))
            sid = json.loads(r.text)['long-url'].split('/')[-1]
            i += 1
            time.sleep(1)
        except:
            print 'done..'
            break
            
    return pd.DataFrame({'user':users, 'tweet':text})
```

```python
df = get_tweetception('433512827177811969')
df.sort_index(ascending=False)
```
